```python
from sklearn.cluster import KMeans
import numpy as np
from sklearn.datasets import load_iris
```

```python

X, y = load_iris(return_X_y=True)
X
```

```python
kmeans = KMeans(n_clusters=3, random_state=2).fit(X)
kmeans.labels_
```

```python
y
```

```python

kmeans.predict([[5.9, 3. , 5.1, 1.8]])
```

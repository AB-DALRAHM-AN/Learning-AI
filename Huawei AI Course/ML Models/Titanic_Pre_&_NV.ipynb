```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("/content/train.csv")
train_data
```

```python
df_test = pd.read_csv("/content/test.csv")
df_test
```

droping some features those are less important as per basic intuition.
making model simple but effective

```python
df_train=train_data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')#inplace
df_train
```

Detrmine inputs and output

```python
inputs = df_train.drop('Survived',axis='columns')
inputs
```

```python
target = df_train.Survived
target
```

Get dummies method pandas

```python
dummies = pd.get_dummies(inputs.Sex  )
dummies.head()
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs
```

I am dropping male column as well because of dummy variable trap theory. One column is enough to repressent male vs female



```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs
```

**checking missing values**

```python
inputs.isnull().sum()
```

**here age has 177 missing values. Filling missing values with meadian value of age column.**

```python
inputs['Age'].median()
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.median())
inputs
```

```python
inputs.isnull().sum()
```

**importing and building model**

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(inputs, target)
```

```python
test_data1 = pd.read_csv("/content/test.csv")
test_data1.head()
```

```python
test_data1.shape
```

**making X_test and y_test**

```python
test_data = test_data1.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
test_data.head()
```

```python
dummies_test = pd.get_dummies(test_data.Sex  )
dummies_test.head()
```

```python
X_test = pd.concat([test_data, dummies_test],axis='columns')
X_test.drop(['Sex','male'],axis='columns',inplace=True)
X_test.head()
```

```python
X_test.isnull().sum()
```

```python
X_test.Fare.mode()
```

```python
X_test.Age = X_test.Age.fillna(X_test.Age.median())
X_test.Fare = X_test.Fare.fillna(X_test.Fare.mean())
X_test.head()
```

```python
X_test[X_test['Fare'].isnull()]
```

```python
y_predicted = model.predict(X_test)
y_predicted
```
